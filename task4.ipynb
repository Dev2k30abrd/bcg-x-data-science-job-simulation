```python
import pandas as pd
import numpy as np

client_data = pd.read_csv("clean_data_after_eda.csv")
price_data = pd.read_csv("price_data (1).csv")
```

```python
client_data['date_activ'] = pd.to_datetime(client_data['date_activ'])
client_data['date_end'] = pd.to_datetime(client_data['date_end'])
client_data['date_modif_prod'] = pd.to_datetime(client_data['date_modif_prod'])
client_data['date_renewal'] = pd.to_datetime(client_data['date_renewal'])

price_data['price_date'] = pd.to_datetime(price_data['price_date'])
```

```python
client_data['contract_duration_days'] = (client_data['date_end'] - client_data['date_activ']).dt.days
client_data['days_since_modification'] = (client_data['date_renewal'] - client_data['date_modif_prod']).dt.days
```

```python
client_data['consumption_diff'] = client_data['cons_12m'] - client_data['forecast_cons_12m']
price_data['price_diff'] = price_data['price_off_peak_var'] - price_data['price_peak_var']
price_grouped = price_data.groupby('id').mean().reset_index()
```

```python
final_data = pd.merge(client_data, price_grouped, on='id', how='left')
```

```python
final_data = final_data.drop(['date_activ','date_end','date_modif_prod','date_renewal'], axis=1)
```

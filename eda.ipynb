```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
client_data = pd.read_csv("client_data (1).csv")
price_data = pd.read_csv("price_data (1).csv")
```

```python
print("Client Data Preview")
print(client_data.head())

print("\nPrice Data Preview")
print(price_data.head())
```

```python
print("Client Data Shape:", client_data.shape)
print("Price Data Shape:", price_data.shape)
```

```python
print("Client Data Columns:")
print(client_data.columns)

print("\nPrice Data Columns:")
print(price_data.columns)
```

```python
print("Client Data Info")
print(client_data.info())

print("\nPrice Data Info")
print(price_data.info())
```

```python
print("Missing Values in Client Data")
print(client_data.isnull().sum())

print("\nMissing Values in Price Data")
print(price_data.isnull().sum())
```

```python
print("Client Data Statistics")
print(client_data.describe())

print("\nPrice Data Statistics")
print(price_data.describe())
```

```python
for col in client_data.columns:
    print(col, ":", client_data[col].nunique())
```

```python
client_data.hist(figsize=(15,10))
plt.show()
```

```python
price_data.hist(figsize=(12,8))
plt.show()
```

```python
client_data['churn'].value_counts().plot(kind='bar')
plt.title("Churn Distribution")
plt.xlabel("Churn")
plt.ylabel("Count")
plt.show()
```

```python
client_data.groupby('churn')['cons_12m'].mean().plot(kind='bar')
plt.title("Average Consumption by Churn")
plt.ylabel("Average cons_12m")
plt.show()
```
